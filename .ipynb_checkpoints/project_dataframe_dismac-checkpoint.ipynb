```python
url = "https://www.dismac.com.bo/categorias/56-climatizacion/106-aire-acondicionado/tipo-split.html"
```

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html = requests.get(url).content
soup = BeautifulSoup(html)
soup 
```

```python
c = soup.find_all("div", class_="product-item-name_mobile_area")
```

```python
c
```

```python
product_name_lst = []
for i in c:
    product = i.find("a")
    product_name_lst.append(product.text)
    
product_name_lst
```

```python
c.find("span",class_="int")
```

```python

price_lst = []
for i in c:
    price = i.find("span",class_="int")
    price_lst.append(price.text)
price_lst
```

```python
url_lst=[]
for x in range(1,3):
    url =  f"https://www.dismac.com.bo/categorias/56-climatizacion/106-aire-acondicionado/tipo-split.html?p={x}"
    url_lst.append(url)
```

```python
lst=[]
name_lst=[]    

for i in url_lst:
    html=requests.get(i).content
    soup=BeautifulSoup(html)
    c = soup.find_all("div", class_="product-item-name_mobile_area")  
    for x in c:
        pri = x.find("span",class_="int")
        lst.append(pri.text)
        name = x.find("a")
        name_lst.append(name.text)
        
        
    
 
        






    
        


```

```python
url_lst
```

```python
c
```

```python
url_lst
```

```python
web_lst= []
for i in url_lst:
    html = requests.get(i).content
    soup = BeautifulSoup(html)
    cont = soup.find_all("div",class_="recent-container list-view")
    for x in cont:
        o=x.find("a", class_="img-product", href =True)
        web_lst.append(o["href"])
        print(o['href'])
      
    
```
